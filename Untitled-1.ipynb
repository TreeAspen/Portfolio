```python
import pandas as pd
import re
import os

# === 1. 更新后的文件列表 (仅保留了你的 -Data.csv 文件路径) ===
# 我已过滤掉了 Metadata 和 Table-Notes 文件，只保留数据文件以避免错误
file_paths = [
    # 2021 Data
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S1901-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S1902-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S1903-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S2001-Data.csv",
    
    # 2022 Data
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S1901-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S1902-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S1903-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S2001-Data.csv",
    
    # 2023 Data
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S1901-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S1902-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S1903-Data.csv",
    r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S2001-Data.csv"
]

def integrate_census_tables_all_types(file_list):
    grouped_files = {}

    # 正则匹配逻辑:
    # 兼容 ACSDP1Y, ACSST5Y 等格式
    pattern = re.compile(r"(ACS[A-Z]{2}[15]Y)(\d{4})\.([A-Z0-9]+)")

    print("正在分类文件...")
    for path in file_list:
        filename = os.path.basename(path)
        
        # 【新增关键步骤】过滤掉 Metadata 和 Notes 文件，只处理 Data.csv
        if "Data.csv" not in filename:
            continue

        match = pattern.search(filename)
        if match:
            dataset_type = match.group(1) # 例如 ACSST5Y
            year = int(match.group(2))    # 例如 2021
            table_id = match.group(3)     # 例如 S1901
            
            # 使用组合键: 数据类型 + 表格编号
            group_key = f"{dataset_type}_{table_id}"
            
            if group_key not in grouped_files:
                grouped_files[group_key] = []
            grouped_files[group_key].append({'year': year, 'path': path})
    
    final_results = {}

    for group_key, files in grouped_files.items():
        # 按年份排序 (2021 -> 2022 -> 2023)
        files.sort(key=lambda x: x['year'])
        
        print(f"\n正在处理组: {group_key} (共 {len(files)} 个年份)")
        
        dfs_to_merge = []
        
        for i, file_info in enumerate(files):
            year = file_info['year']
            path = file_info['path']
            
            try:
                # 读取数据 (Census CSV通常第二行是描述，所以 header=0 读取第一行作为列名)
                df = pd.read_csv(path, header=0, low_memory=False)
                
                if not df.empty:
                    # 1. 添加 YEAR 列标记年份
                    df['YEAR'] = year
                    
                    if i == 0:
                        # 【如果是该组的第一个文件 (如 2021)】：
                        # 保留 Index 0 (描述行: Estimate!!Total...)。
                        # 并将描述行的 YEAR 列改为字符串 "Year" 以保持格式整洁
                        if 'YEAR' in df.columns:
                            year_col_idx = df.columns.get_loc('YEAR')
                            df.iloc[0, year_col_idx] = 'Year'
                            
                        print(f"  - 已读取 {year} 年数据 (保留描述行)")
                        
                    else:
                        # 【如果是后续文件 (2022, 2023)】：
                        # 删除 Index 0 (描述行)，只保留纯数据，避免重复
                        df = df.drop(index=0)
                        print(f"  - 已读取 {year} 年数据 (删除描述行)")
                    
                    # 调整列顺序：把 YEAR 放在第一列方便查看
                    cols = df.columns.tolist()
                    cols = cols[-1:] + cols[:-1]
                    df = df[cols]
                    
                    dfs_to_merge.append(df)

            except Exception as e:
                print(f"  ! 错误: {path} - {e}")

        # 合并数据
        if dfs_to_merge:
            merged_df = pd.concat(dfs_to_merge, ignore_index=True)
            final_results[group_key] = merged_df
            print(f"  > 组 {group_key} 合并完成! 总行数: {len(merged_df)}")
    
    return final_results

# === 主执行块 ===
if __name__ == "__main__":
    results = integrate_census_tables_all_types(file_paths)
    
    # 输出目录 (保持和你输入目录一致)
    output_dir = r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data"
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    print("\n=== 开始导出文件 ===")
    for group_key, df in results.items():
        # 文件名例如: Merged_ACSST5Y_S1901.csv
        output_filename = f"Merged_{group_key}.csv"
        save_path = os.path.join(output_dir, output_filename)
        
        df.to_csv(save_path, index=False)
        print(f"文件已保存至: {save_path}")
```

```python
import pandas as pd
import os

# ---------------------------------------------------------
# 1. 配置文件路径 (根据你提供的列表整理)
# ---------------------------------------------------------
# 为了代码简洁，我们构建一个结构化的字典
# 键格式: (Year, Type) -> { 'data': path, 'meta': path }
# 我们主要选取三个核心文件来代表 SES 和 Housing Market:
# - DP05: 人口 (Population)
# - S1901: 收入 (Income)
# - DP04: 住房特征 (Housing Value & Rent)

files_map = {
    2021: {
        'DP05': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2021.DP05-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2021.DP05-Column-Metadata.csv"
        },
        'S1901': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S1901-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2021.S1901-Column-Metadata.csv"
        },
        'DP04': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2021.DP04-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2021.DP04-Column-Metadata.csv"
        }
    },
    2022: {
        'DP05': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2022.DP05-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2022.DP05-Column-Metadata.csv"
        },
        'S1901': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S1901-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2022.S1901-Column-Metadata.csv"
        },
        'DP04': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2022.DP04-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2022.DP04-Column-Metadata.csv"
        }
    },
    2023: {
        'DP05': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2023.DP05-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2023.DP05-Column-Metadata.csv"
        },
        'S1901': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S1901-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSST5Y2023.S1901-Column-Metadata.csv"
        },
        'DP04': {
            'data': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2023.DP04-Data.csv",
            'meta': r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACSDP5Y2023.DP04-Column-Metadata.csv"
        }
    }
}

# ---------------------------------------------------------
# 2. 核心函数
# ---------------------------------------------------------
def clean_county_name(name):
    """标准化县名：转大写，去除 ' County' 结尾"""
    if not isinstance(name, str): return str(name)
    # 大部分 ACS 格式为 "Albany County, New York"
    clean = name.split(',')[0] # 取逗号前部分 "Albany County"
    clean = clean.replace(' County', '').strip().upper() # "ALBANY"
    return clean

def find_column_id(meta_path, keyword):
    """从 Metadata 文件中根据关键词查找列 ID"""
    try:
        df_meta = pd.read_csv(meta_path)
        # 假设 Metadata 列名为 'Column Name' 和 'Label'
        # 我们寻找 Label 中包含 keyword 的第一行
        # 注意：ACS Label 往往很长，如 "Estimate!!Total"
        match = df_meta[df_meta['Label'].str.contains(keyword, case=False, na=False)]
        if not match.empty:
            # 返回第一个匹配的列名
            col_id = match.iloc[0]['Column Name']
            print(f"  [Found] '{keyword}' -> {col_id} in {os.path.basename(meta_path)}")
            return col_id
        else:
            print(f"  [Warning] Not found '{keyword}' in {os.path.basename(meta_path)}")
            return None
    except Exception as e:
        print(f"  [Error] Reading metadata {meta_path}: {e}")
        return None

def process_acs_year(year, file_dict):
    print(f"\n--- Processing Year: {year} ---")
    
    # 1. DP05: Population (总人口)
    # 通常关键词: "Estimate!!Total" 且在最前面, 或者直接找 "Total population"
    # DP05 的第一个变量通常就是 Total Population (DP05_0001E)
    path_dp05_data = file_dict['DP05']['data']
    # 直接读取 Data，假设第二行(header=1)是描述，但我们用 Column ID 更稳
    # 这里我们直接指定常见 ID，如果找不到再打印 Warning
    # DP05_0001E = Total Population
    
    df_dp05 = pd.read_csv(path_dp05_data, dtype={'GEO_ID': str}, low_memory=False)
    # 清洗：ACS 数据通常第二行是描述，第一行是 ID。如果读进来第一行是描述，需要处理
    if df_dp05.iloc[0]['NAME'] == 'Geographic Area Name':
        df_dp05 = df_dp05.iloc[1:] # 删除描述行
    
    # 提取 Population
    # 尝试列名: DP05_0001E (如果是 5年期数据，通常是这个)
    pop_col = 'DP05_0001E'
    if pop_col not in df_dp05.columns:
        # 如果列名不对，这里可以加自动查找逻辑，但 DP05_0001E 很标准
        print(f"Warning: {pop_col} not found in DP05. Check columns.")
    
    df_pop = df_dp05[['NAME', pop_col]].copy()
    df_pop.rename(columns={pop_col: 'Population'}, inplace=True)
    
    # 2. S1901: Income (中位数家庭收入)
    # 关键词: "Median income (dollars)!!Estimate!!Households"
    # 代码通常是 S1901_C01_012E (但会有变化，我们用 Metadata 查一下稳妥)
    path_s1901_meta = file_dict['S1901']['meta']
    path_s1901_data = file_dict['S1901']['data']
    
    # 查找 "Median income (dollars)" 且针对 "Households"
    # 简化：找包含 "Median income (dollars)" 的第一个列
    inc_col = find_column_id(path_s1901_meta, "Median income \(dollars\)")
    
    df_s1901 = pd.read_csv(path_s1901_data, dtype={'GEO_ID': str}, low_memory=False)
    if df_s1901.iloc[0]['NAME'] == 'Geographic Area Name': df_s1901 = df_s1901.iloc[1:]
    
    if inc_col and inc_col in df_s1901.columns:
        df_inc = df_s1901[['NAME', inc_col]].copy()
        df_inc.rename(columns={inc_col: 'Median_Household_Income'}, inplace=True)
    else:
        print("Warning: Could not extract Income column.")
        df_inc = df_s1901[['NAME']].copy() # 空占位
        df_inc['Median_Household_Income'] = None

    # 3. DP04: Housing (中位数房价 & 租金)
    # 房价: "Median (dollars)" under Owner-occupied units -> DP04_0089E (往往)
    # 租金: "Median (dollars)" under Gross rent -> DP04_0134E (往往)
    path_dp04_meta = file_dict['DP04']['meta']
    path_dp04_data = file_dict['DP04']['data']
    
    # 查找关键词 (Metadata 中 Label 往往很长，这是个挑战)
    # 手动指定最常见的 5-Year ACS 代码以提高成功率，若你需要极高精度，请手动核对 Metadata
    # 2021-2023 ACS 5-Year DP04:
    # DP04_0089E: Estimate!!VALUE!!Owner-occupied units!!Median (dollars)
    # DP04_0134E: Estimate!!GROSS RENT!!Occupied units paying rent!!Median (dollars)
    
    val_col = 'DP04_0089E' 
    rent_col = 'DP04_0134E'
    
    df_dp04 = pd.read_csv(path_dp04_data, dtype={'GEO_ID': str}, low_memory=False)
    if df_dp04.iloc[0]['NAME'] == 'Geographic Area Name': df_dp04 = df_dp04.iloc[1:]
    
    cols_to_keep = ['NAME']
    rename_map = {}
    
    if val_col in df_dp04.columns:
        cols_to_keep.append(val_col)
        rename_map[val_col] = 'Median_Home_Value'
    
    if rent_col in df_dp04.columns:
        cols_to_keep.append(rent_col)
        rename_map[rent_col] = 'Median_Rent'
        
    df_house = df_dp04[cols_to_keep].copy()
    df_house.rename(columns=rename_map, inplace=True)
    
    # 4. 合并当前年份
    # 基准: Population 表
    df_merged = df_pop
    df_merged = pd.merge(df_merged, df_inc, on='NAME', how='left')
    df_merged = pd.merge(df_merged, df_house, on='NAME', how='left')
    
    # 添加年份和清洗县名
    df_merged['year'] = year
    df_merged['clean_county'] = df_merged['NAME'].apply(clean_county_name)
    
    # 转换数值列 (去除可能的 '-' 或 '(X)' 或 'N' 等非数值字符)
    num_cols = ['Population', 'Median_Household_Income', 'Median_Home_Value', 'Median_Rent']
    for col in num_cols:
        if col in df_merged.columns:
            df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')
            
    return df_merged

# ---------------------------------------------------------
# 3. 主执行流程
# ---------------------------------------------------------
all_years_data = []

for year in [2021, 2022, 2023]:
    df_year = process_acs_year(year, files_map[year])
    all_years_data.append(df_year)

# 纵向合并
final_acs_panel = pd.concat(all_years_data, ignore_index=True)

# 筛选最终列
final_cols = ['clean_county', 'year', 'Population', 'Median_Household_Income', 'Median_Home_Value', 'Median_Rent']
# 确保列存在
final_cols = [c for c in final_cols if c in final_acs_panel.columns]
final_acs_panel = final_acs_panel[final_cols]

# 预览
print("\n=== ACS Panel Data Preview ===")
print(final_acs_panel.head())
print(final_acs_panel.describe())

# 导出
output_file = r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\ACS_SES_Housing_Panel_2021_2023.csv"
final_acs_panel.to_csv(output_file, index=False)
print(f"\nSuccessfully saved ACS Panel Data to: {output_file}")
```

```python
import pandas as pd
import os

# ---------------------------------------------------------
# 1. 配置文件路径 (已添加 S2001)
# ---------------------------------------------------------
# 根目录
base_dir = r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data"

files_map = {
    2021: {
        'DP05':  os.path.join(base_dir, "ACSDP5Y2021.DP05-Data.csv"),
        'S1901': os.path.join(base_dir, "ACSST5Y2021.S1901-Data.csv"),
        'DP04':  os.path.join(base_dir, "ACSDP5Y2021.DP04-Data.csv"),
        'S2001': os.path.join(base_dir, "ACSST5Y2021.S2001-Data.csv")  # 新增
    },
    2022: {
        'DP05':  os.path.join(base_dir, "ACSDP5Y2022.DP05-Data.csv"),
        'S1901': os.path.join(base_dir, "ACSST5Y2022.S1901-Data.csv"),
        'DP04':  os.path.join(base_dir, "ACSDP5Y2022.DP04-Data.csv"),
        'S2001': os.path.join(base_dir, "ACSST5Y2022.S2001-Data.csv")  # 新增
    },
    2023: {
        'DP05':  os.path.join(base_dir, "ACSDP5Y2023.DP05-Data.csv"),
        'S1901': os.path.join(base_dir, "ACSST5Y2023.S1901-Data.csv"),
        'DP04':  os.path.join(base_dir, "ACSDP5Y2023.DP04-Data.csv"),
        'S2001': os.path.join(base_dir, "ACSST5Y2023.S2001-Data.csv")  # 新增
    }
}

# ---------------------------------------------------------
# 2. 核心处理函数
# ---------------------------------------------------------
def clean_county_name(name):
    """标准化县名：转大写，去除 ' County' 结尾"""
    if not isinstance(name, str): return str(name)
    # 处理 "Albany County, New York" -> "ALBANY"
    clean = name.split(',')[0] 
    clean = clean.replace(' County', '').strip().upper()
    return clean

def read_acs_file(file_path, cols_mapping):
    """
    通用读取函数
    :param file_path: CSV路径
    :param cols_mapping: 字典 { 'ACS列代码': '新列名' }
    """
    if not os.path.exists(file_path):
        print(f"  [Error] 文件不存在: {file_path}")
        return pd.DataFrame()

    try:
        # 读取数据，指定 GEO_ID 为字符串防止前导0丢失
        df = pd.read_csv(file_path, dtype={'GEO_ID': str}, low_memory=False)
        
        # 清洗: 如果第二行是描述行 (Name列是 Geographic Area Name)，则删除
        if not df.empty and df.iloc[0]['NAME'] == 'Geographic Area Name':
            df = df.iloc[1:]
        
        # 准备提取数据
        # 总是保留 NAME 用作合并键
        keep_cols = ['NAME']
        rename_dict = {}
        
        for acs_code, new_name in cols_mapping.items():
            if acs_code in df.columns:
                keep_cols.append(acs_code)
                rename_dict[acs_code] = new_name
            else:
                print(f"  [Warning] 未找到列 {acs_code} ({new_name}) in {os.path.basename(file_path)}")
        
        # 筛选并重命名
        df_subset = df[keep_cols].copy()
        df_subset.rename(columns=rename_dict, inplace=True)
        return df_subset

    except Exception as e:
        print(f"  [Error] 读取 {os.path.basename(file_path)} 失败: {e}")
        return pd.DataFrame()

def process_acs_year(year, paths):
    print(f"\n--- 正在处理年份: {year} ---")
    
    # === A. 配置需要提取的列 (ACS 5-Year 标准代码) ===
    # 你可以根据 metadata 核对这些代码，这些是通常的标准代码
    
    # 1. DP05: 人口
    map_dp05 = {
        'DP05_0001E': 'Population' 
    }
    
    # 2. S1901: 家庭收入 (Median Household Income)
    map_s1901 = {
        'S1901_C01_012E': 'Median_Household_Income' 
    }
    
    # 3. DP04: 住房 (Value & Rent)
    map_dp04 = {
        'DP04_0089E': 'Median_Home_Value', # Owner-occupied units value
        'DP04_0134E': 'Median_Rent'        # Gross rent
    }
    
    # 4. S2001: 个人收入 (Median Earnings)
    map_s2001 = {
        'S2001_C01_002E': 'Median_Earnings' # Total !! Estimate !! Median earnings
    }

    # === B. 读取各个文件 ===
    df_pop   = read_acs_file(paths['DP05'], map_dp05)
    df_inc   = read_acs_file(paths['S1901'], map_s1901)
    df_house = read_acs_file(paths['DP04'], map_dp04)
    df_earn  = read_acs_file(paths['S2001'], map_s2001)

    # === C. 合并数据 ===
    # 以 Population 表为基准 (它通常最全)
    if df_pop.empty:
        print("  [Error] 跳过该年份，因为人口数据读取失败")
        return pd.DataFrame()

    merged = df_pop
    
    # 依次合并 (Left Join)
    if not df_inc.empty:
        merged = pd.merge(merged, df_inc, on='NAME', how='left')
    if not df_house.empty:
        merged = pd.merge(merged, df_house, on='NAME', how='left')
    if not df_earn.empty:
        merged = pd.merge(merged, df_earn, on='NAME', how='left')

    # === D. 清洗 ===
    merged['year'] = year
    merged['clean_county'] = merged['NAME'].apply(clean_county_name)
    
    # 转换数值 (处理 '-', '(X)', 'N' 等)
    numeric_cols = ['Population', 'Median_Household_Income', 'Median_Home_Value', 'Median_Rent', 'Median_Earnings']
    for col in numeric_cols:
        if col in merged.columns:
            merged[col] = pd.to_numeric(merged[col], errors='coerce')

    return merged

# ---------------------------------------------------------
# 3. 主执行流程
# ---------------------------------------------------------
if __name__ == "__main__":
    all_years_data = []

    for year in [2021, 2022, 2023]:
        df_year = process_acs_year(year, files_map[year])
        if not df_year.empty:
            all_years_data.append(df_year)

    if all_years_data:
        # 纵向合并所有年份
        final_acs_panel = pd.concat(all_years_data, ignore_index=True)

        # 整理列顺序
        target_cols = ['clean_county', 'year', 'Population', 'Median_Household_Income', 'Median_Earnings', 'Median_Home_Value', 'Median_Rent']
        # 只保留存在的列
        final_cols = [c for c in target_cols if c in final_acs_panel.columns]
        final_acs_panel = final_acs_panel[final_cols]

        # 预览
        print("\n=== 最终面板数据预览 (前 5 行) ===")
        print(final_acs_panel.head())
        print("\n=== 数据描述统计 ===")
        print(final_acs_panel.describe())

        # 导出
        output_file = os.path.join(base_dir, "ACS_SES_Housing_Panel_2021_2023.csv")
        final_acs_panel.to_csv(output_file, index=False)
        print(f"\n文件成功保存至: {output_file}")
    else:
        print("未生成任何数据，请检查文件路径。")
```

```python
import pandas as pd
import os
import numpy as np

# ================= 配置路径 =================
INPUT_DIR = r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\filtered_2021_2023"

FILE_DEC  = os.path.join(INPUT_DIR, "Filtered_1_Declarations_NY_2021-2023.csv")
FILE_PA   = os.path.join(INPUT_DIR, "Filtered_2_PublicAssistance_NY_2021-2023.csv")
FILE_IA   = os.path.join(INPUT_DIR, "Filtered_3_IA_Owners_NY_2021-2023.csv")
FILE_HMGP = os.path.join(INPUT_DIR, "Filtered_4_HMGP_Mitigation_NY_2021-2023.csv")

OUTPUT_FILE = os.path.join(INPUT_DIR, "Final_Merged_Dataset_NY_Unified_Count.csv")

# ================= 清洗工具函数 =================
def clean_disaster_number(series):
    return pd.to_numeric(series, errors='coerce').fillna(0).astype(int)

def clean_county(series):
    if series.empty: return series
    return (series.astype(str).str.upper()
            .str.replace(r"\(COUNTY\)", "", regex=True)
            .str.replace(" COUNTY", "", regex=False)
            .str.strip())

# ================= 1. 读取并清洗数据 =================
print("正在读取数据...")
df_dec  = pd.read_csv(FILE_DEC, dtype=str, low_memory=False)
df_pa   = pd.read_csv(FILE_PA, dtype=str, low_memory=False)
df_ia   = pd.read_csv(FILE_IA, dtype=str, low_memory=False)
df_hmgp = pd.read_csv(FILE_HMGP, dtype=str, low_memory=False)

for df in [df_dec, df_pa, df_ia, df_hmgp]:
    df.columns = df.columns.str.upper().str.strip()

# --- 标准化关联键 ---
print("正在标准化关联键...")
df_dec["DISASTER_KEY"] = clean_disaster_number(df_dec["DISASTERNUMBER"])
df_dec["COUNTY_KEY"]   = clean_county(df_dec.get("DESIGNATEDAREA", df_dec.get("COUNTY")))

df_pa["DISASTER_KEY"] = clean_disaster_number(df_pa["DISASTERNUMBER"])
df_pa["COUNTY_KEY"]   = clean_county(df_pa["COUNTY"])

df_ia["DISASTER_KEY"] = clean_disaster_number(df_ia["DISASTERNUMBER"])
df_ia["COUNTY_KEY"]   = clean_county(df_ia["COUNTY"])

df_hmgp["DISASTER_KEY"] = clean_disaster_number(df_hmgp["DISASTERNUMBER"])
df_hmgp["COUNTY_KEY"]   = clean_county(df_hmgp["COUNTY"])

# ================= 2. 统一 HMGP 计数列 (核心修改) =================
print("正在计算 HMGP_COUNT...")

# 1. 确定金额列 (优先找 Actual, 没有找 Project)
hmgp_amount_target = next((c for c in ["ACTUALAMOUNTPAID", "ACTUALAMOUNT", "PROJECTAMOUNT"] if c in df_hmgp.columns), None)
df_hmgp["HMGP_AMOUNT_CLEAN"] = pd.to_numeric(df_hmgp[hmgp_amount_target], errors='coerce').fillna(0) if hmgp_amount_target else 0

# 2. 确定数量列 (NUMBEROFPROPERTIES)
# 如果这一列存在，就用它；如果不存在或为空，则默认为1（代表该项目至少涉及1处房产）
if "NUMBEROFPROPERTIES" in df_hmgp.columns:
    df_hmgp["HMGP_PROPS_RAW"] = pd.to_numeric(df_hmgp["NUMBEROFPROPERTIES"], errors='coerce')
    # 填充逻辑：如果有项目记录但数量为空，通常默认为1
    df_hmgp["HMGP_PROPS_CLEAN"] = df_hmgp["HMGP_PROPS_RAW"].fillna(1)
else:
    df_hmgp["HMGP_PROPS_CLEAN"] = 1

# ================= 3. 聚合数据 =================
print("正在聚合...")

# PA: 金额 & 项目数
pa_amount_col = next((c for c in ["PROJECTAMOUNT", "TOTALOBLIGATED"] if c in df_pa.columns), None)
df_pa["PA_AMT"] = pd.to_numeric(df_pa[pa_amount_col], errors='coerce').fillna(0) if pa_amount_col else 0

pa_agg = df_pa.groupby(["DISASTER_KEY", "COUNTY_KEY"]).agg({
    "PA_AMT": "sum",
    "DISASTER_KEY": "count"
}).rename(columns={"PA_AMT": "PA_Total_Amount", "DISASTER_KEY": "PA_Project_Count"}).reset_index()

# IA: 金额
ia_amount_col = "TOTALAPPROVEDIHPAMOUNT"
df_ia["IA_AMT"] = pd.to_numeric(df_ia[ia_amount_col], errors='coerce').fillna(0) if ia_amount_col in df_ia.columns else 0

ia_agg = df_ia.groupby(["DISASTER_KEY", "COUNTY_KEY"])["IA_AMT"].sum().reset_index(name="IA_Total_Amount")

# HMGP: 这里的 HMGP_COUNT 就是你要的“合并列”
hmgp_agg = df_hmgp.groupby(["DISASTER_KEY", "COUNTY_KEY"]).agg({
    "HMGP_AMOUNT_CLEAN": "sum",
    "HMGP_PROPS_CLEAN": "sum"  # 这里把所有房产数加总，不再分 Buyout 或 Project Count
}).rename(columns={
    "HMGP_AMOUNT_CLEAN": "HMGP_Total_Amount",
    "HMGP_PROPS_CLEAN": "HMGP_COUNT" # <--- 最终合并后的单一计数列
}).reset_index()

# ================= 4. 合并与保存 =================
print("正在合并主表...")

# Declaration 骨架
base_cols = ["DISASTER_KEY", "COUNTY_KEY", "STATE", "DECLARATIONDATE", "INCIDENTTYPE", "DECLARATIONTITLE"]
base_df = df_dec[base_cols].drop_duplicates(subset=["DISASTER_KEY", "COUNTY_KEY"])

# 合并
final_df = base_df.merge(pa_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")
final_df = final_df.merge(ia_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")
final_df = final_df.merge(hmgp_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")

# 填充0
for col in ["PA_Total_Amount", "PA_Project_Count", "IA_Total_Amount", "HMGP_Total_Amount", "HMGP_COUNT"]:
    final_df[col] = final_df[col].fillna(0)

# 总金额
final_df["Total_FEMA_Assistance"] = (
    final_df["PA_Total_Amount"] + 
    final_df["IA_Total_Amount"] + 
    final_df["HMGP_Total_Amount"]
)

final_df.to_csv(OUTPUT_FILE, index=False)
print("="*30)
print(f"处理完成！HMGP 计数已合并为单列: HMGP_COUNT")
print(f"文件保存至: {OUTPUT_FILE}")
print("HMGP 数据预览:")
print(final_df[final_df["HMGP_COUNT"] > 0][["DISASTER_KEY", "COUNTY_KEY", "HMGP_Total_Amount", "HMGP_COUNT"]].head())
```

```python
import pandas as pd
import os
import numpy as np

# ================= 配置路径 =================
# 输入文件夹 (对应你上一步脚本的输出)
INPUT_DIR = r"C:\Users\sy4254\Desktop\Casaul Final\Origin_census_data\filtered_NY_AllTime_Natural"

# 输入文件路径
FILE_DEC  = os.path.join(INPUT_DIR, "Filtered_1_Declarations_NY_AllTime.csv")
FILE_PA   = os.path.join(INPUT_DIR, "Filtered_2_PublicAssistance_NY_AllTime.csv")
FILE_IA   = os.path.join(INPUT_DIR, "Filtered_3_IA_Owners_NY_AllTime.csv")
FILE_HMGP = os.path.join(INPUT_DIR, "Filtered_4_HMGP_Mitigation_NY_AllTime.csv")

# 输出文件
OUTPUT_FILE = os.path.join(INPUT_DIR, "Final_Merged_Dataset_NY_AllTime_Natural_Unified.csv")

# ================= 清洗工具函数 =================
def clean_disaster_number(series):
    """统一灾害编号格式 (移除小数点, 转整数)"""
    return pd.to_numeric(series, errors='coerce').fillna(0).astype(int)

def clean_county(series):
    """统一县名格式 (大写, 去除后缀)"""
    if series.empty: return series
    return (series.astype(str).str.upper()
            .str.replace(r"\(COUNTY\)", "", regex=True)
            .str.replace(" COUNTY", "", regex=False)
            .str.strip())

# ================= 1. 读取并预处理数据 =================
print("正在读取筛选后的数据...")

# 读取 (全部按字符串读取以防格式丢失)
df_dec  = pd.read_csv(FILE_DEC, dtype=str, low_memory=False)
df_pa   = pd.read_csv(FILE_PA, dtype=str, low_memory=False)
df_ia   = pd.read_csv(FILE_IA, dtype=str, low_memory=False)
df_hmgp = pd.read_csv(FILE_HMGP, dtype=str, low_memory=False)

# 列名标准化
for df in [df_dec, df_pa, df_ia, df_hmgp]:
    df.columns = df.columns.str.upper().str.strip()

print("正在标准化关联键 (Disaster Key & County Key)...")

# --- 生成统一的关联键 ---
# Declarations
df_dec["DISASTER_KEY"] = clean_disaster_number(df_dec["DISASTERNUMBER"])
target_county_col = "DESIGNATEDAREA" if "DESIGNATEDAREA" in df_dec.columns else "COUNTY"
df_dec["COUNTY_KEY"] = clean_county(df_dec[target_county_col])

# PA
df_pa["DISASTER_KEY"] = clean_disaster_number(df_pa["DISASTERNUMBER"])
df_pa["COUNTY_KEY"]   = clean_county(df_pa["COUNTY"])

# IA
df_ia["DISASTER_KEY"] = clean_disaster_number(df_ia["DISASTERNUMBER"])
df_ia["COUNTY_KEY"]   = clean_county(df_ia["COUNTY"])

# HMGP
df_hmgp["DISASTER_KEY"] = clean_disaster_number(df_hmgp["DISASTERNUMBER"])
df_hmgp["COUNTY_KEY"]   = clean_county(df_hmgp["COUNTY"])

# ================= 2. 字段智能清洗与计算 =================
print("正在清洗金额与计数列...")

# --- 2.1 PA (公共援助) ---
# 寻找金额列
pa_amt_col = next((c for c in ["PROJECTAMOUNT", "TOTALOBLIGATED", "FEDERALSHAREOBLIGATED"] if c in df_pa.columns), None)
df_pa["PA_AMT_CLEAN"] = pd.to_numeric(df_pa[pa_amt_col], errors='coerce').fillna(0) if pa_amt_col else 0

# --- 2.2 IA (个人援助) ---
ia_amt_col = "TOTALAPPROVEDIHPAMOUNT"
df_ia["IA_AMT_CLEAN"] = pd.to_numeric(df_ia[ia_amt_col], errors='coerce').fillna(0) if ia_amt_col in df_ia.columns else 0

# --- 2.3 HMGP (减灾) - 核心逻辑 ---
# A. 金额列: 优先找 ACTUALAMOUNT (实际金额), 其次 PROJECTAMOUNT (项目金额)
hmgp_amt_col = next((c for c in ["ACTUALAMOUNT", "ACTUALAMOUNTPAID", "PROJECTAMOUNT"] if c in df_hmgp.columns), None)
df_hmgp["HMGP_AMT_CLEAN"] = pd.to_numeric(df_hmgp[hmgp_amt_col], errors='coerce').fillna(0) if hmgp_amt_col else 0

# B. 数量列: 统一使用 NUMBEROFPROPERTIES (房产数)
# 如果该列存在且有值，则用它；如果为空或不存在，默认为 1 (表示至少涉及1处资产)
if "NUMBEROFPROPERTIES" in df_hmgp.columns:
    df_hmgp["HMGP_PROPS_RAW"] = pd.to_numeric(df_hmgp["NUMBEROFPROPERTIES"], errors='coerce')
    df_hmgp["HMGP_PROPS_CLEAN"] = df_hmgp["HMGP_PROPS_RAW"].fillna(1)
else:
    df_hmgp["HMGP_PROPS_CLEAN"] = 1

# C. Buyout (收购) 识别
# 寻找动作描述列
action_col = next((c for c in ["PROPERTYACTION", "PROJECTTYPE", "ACTION"] if c in df_hmgp.columns), None)

if action_col:
    # 查找包含 "ACQUISITION" (收购) 的记录
    is_buyout = df_hmgp[action_col].astype(str).str.upper().str.contains("ACQUISITION", na=False)
    
    # 提取 Buyout 的金额和数量
    df_hmgp["Buyout_Amount"] = np.where(is_buyout, df_hmgp["HMGP_AMT_CLEAN"], 0)
    df_hmgp["Buyout_Count"]  = np.where(is_buyout, df_hmgp["HMGP_PROPS_CLEAN"], 0)
else:
    print(" [警告] HMGP 中未找到 PROPERTYACTION 列，无法单独统计 Buyout。")
    df_hmgp["Buyout_Amount"] = 0
    df_hmgp["Buyout_Count"] = 0

# ================= 3. 数据聚合 (Aggregation) =================
print("正在聚合数据...")

# PA 聚合
pa_agg = df_pa.groupby(["DISASTER_KEY", "COUNTY_KEY"]).agg({
    "PA_AMT_CLEAN": "sum",
    "DISASTER_KEY": "count" # 这里的 count 是项目数量 (行数)
}).rename(columns={"PA_AMT_CLEAN": "PA_Total_Amount", "DISASTER_KEY": "PA_Project_Count"}).reset_index()

# IA 聚合
ia_agg = df_ia.groupby(["DISASTER_KEY", "COUNTY_KEY"]).agg({
    "IA_AMT_CLEAN": "sum"
}).rename(columns={"IA_AMT_CLEAN": "IA_Total_Amount"}).reset_index()

# HMGP 聚合
# 注意: HMGP_COUNT 现在是所有项目涉及的房产总数 (包含 Buyout 和非 Buyout)
hmgp_agg = df_hmgp.groupby(["DISASTER_KEY", "COUNTY_KEY"]).agg({
    "HMGP_AMT_CLEAN": "sum",
    "HMGP_PROPS_CLEAN": "sum",
    "Buyout_Amount": "sum",
    "Buyout_Count": "sum"
}).rename(columns={
    "HMGP_AMT_CLEAN": "HMGP_Total_Amount",
    "HMGP_PROPS_CLEAN": "HMGP_COUNT",  # <-- 统一的房产计数列
    "Buyout_Amount": "Buyout_Total_Amount",
    "Buyout_Count": "Buyout_Property_Count"
}).reset_index()

# ================= 4. 最终合并 (Merge) =================
print("正在合并主表...")

# 构建骨架 (只保留 Declarations 中的关键信息)
base_cols = ["DISASTER_KEY", "COUNTY_KEY", "STATE", "DECLARATIONDATE", "INCIDENTTYPE", "DECLARATIONTITLE"]
# 确保列存在再选取
base_cols = [c for c in base_cols if c in df_dec.columns]
base_df = df_dec[base_cols].drop_duplicates(subset=["DISASTER_KEY", "COUNTY_KEY"])

# 依次左连接
final_df = base_df.merge(pa_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")
final_df = final_df.merge(ia_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")
final_df = final_df.merge(hmgp_agg, on=["DISASTER_KEY", "COUNTY_KEY"], how="left")

# ================= 5. 收尾工作 =================
# 填充空值为 0
fill_cols = [
    "PA_Total_Amount", "PA_Project_Count",
    "IA_Total_Amount",
    "HMGP_Total_Amount", "HMGP_COUNT",
    "Buyout_Total_Amount", "Buyout_Property_Count"
]

for col in fill_cols:
    if col in final_df.columns:
        final_df[col] = final_df[col].fillna(0)

# 计算总援助金额 (Total Assistance)
final_df["Total_FEMA_Assistance"] = (
    final_df["PA_Total_Amount"] + 
    final_df["IA_Total_Amount"] + 
    final_df["HMGP_Total_Amount"]
)

# 保存
final_df.to_csv(OUTPUT_FILE, index=False)

print("=" * 30)
print(f"合并完成！(全时间段 - 自然灾害)")
print(f"输出文件路径: {OUTPUT_FILE}")
print("-" * 30)
print("数据概览 (非零数据前5行):")
preview = final_df[final_df["Total_FEMA_Assistance"] > 0][["DISASTER_KEY", "COUNTY_KEY", "INCIDENTTYPE", "Total_FEMA_Assistance", "HMGP_COUNT"]]
print(preview.head())
```
